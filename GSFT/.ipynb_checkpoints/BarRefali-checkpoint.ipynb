```python
def bar_create(ohlc,what = 'transfers_volume_sum',th=5):
    df = ohlc.copy()
    group = pd.Series(index = df.index, dtype = float, name = 'group')
    N = len(group)

    t = ohlc[what] #100*np.log(ohlc['Close']/ohlc['Open'])
    ts = 0
    idx = []

    for i, x in enumerate(t):
        ts += x
        if np.abs(ts) > th:
            idx.append(i)
            ts = 0.0
            continue

    group.iloc[idx]  = idx
    group.fillna(method = 'bfill',inplace = True)
    group.fillna(method = 'ffill',inplace = True)

    df.loc[:,'group'] = group
    df.loc[:,'Time'] = df.index

    fTime = df.groupby('group')['Time'].first()
    lTime = df.groupby('group')['Time'].last()
    Time_ = pd.concat([fTime,lTime],axis = 1)
    Time_.columns = ['open','close']
    Time_.index = Time_.close
    Time_.loc[:,'dt'] = (Time_.close-Time_.open).dt.total_seconds()


    Open = df.groupby('group')['Open'].first()
    High = df.groupby('group')['High'].max()
    Low = df.groupby('group')['Low'].min()
    Close = df.groupby('group')['Close'].last()

    newOhlc = pd.concat([Open,High,Low,Close],axis = 1)
    newOhlc.index = lTime
    return newOhlc,group

def check_direction(R):
    pos = (R>0) & (R.shift(1)>0)
    neg = (R<=0) & (R.shift(1)<=0)

    mask = pos | neg
    return mask.sum()/mask.count()
```

```python
Urls['transfers_volume_sum'] = 'https://api.glassnode.com/v1/metrics/transactions/transfers_volume_sum'
```

```python
ohlc = get_glassnode_price()
ohlc = ohlc.rename(columns = {'o':'Open','h':'High','l':'Low','c':'Close'})

sam_list = ['mvrv',
            'sopr',
            'rhodl_ratio',
            'utxo_created_value_sum',
            'active_count',
            'puell_multiple',
            'count',
            'mvrv_z_score',
            'deltacap_usd',
            'marketcap_usd',
            'transfers_volume_exchanges_net',
            'transfers_volume_sum']

features = get_glassnode_data(sam_list,Urls)

ohlc = pd.concat([ohlc,features],axis=1)
ohlc = ohlc.loc[:'2018']
ohlc = ohlc.dropna()
ohlc['Ret'] = 100*np.log(ohlc['Close']/ohlc['Open'])
```

```python
newOhlc,group = bar_create(ohlc,what = 'transfers_volume_sum',th = 7.5*1e6)
#newOhlc,group = bar_create(ohlc,what = 'Ret',th = 5)
```

```python
f,a = plt.subplots(nrows = 2)
mpf.plot(ohlc.loc['2018'],type='candle',ax = a[0])
mpf.plot(newOhlc.loc['2018'],type='candle',ax = a[1])
```

```python
ohlc['group'] = group
MuFeatures = ohlc.groupby('group')[sam_list].mean()
StFeatures = ohlc.groupby('group')[sam_list].std()
MuFeatures.index = newOhlc.index
StFeatures.index = newOhlc.index
ZtFeatures = MuFeatures/StFeatures
```

```python
R = 100*np.log(newOhlc['Close']/newOhlc['Open'])
MuFeatures['R']=R
StFeatures['R']=R
ZtFeatures['R']=R

MuFeatures['LR']=R.shift(1)
StFeatures['LR']=R.shift(1)
ZtFeatures['LR']=R.shift(1)

MuFeatures.insert(0, 'R', MuFeatures.pop('R'))
StFeatures.insert(0, 'R', StFeatures.pop('R'))
ZtFeatures.insert(0, 'R', ZtFeatures.pop('R'))

MuFeatures.insert(1, 'LR', MuFeatures.pop('LR'))
StFeatures.insert(1, 'LR', StFeatures.pop('LR'))
ZtFeatures.insert(1, 'LR', ZtFeatures.pop('LR'))

MuFeatures=MuFeatures.dropna()
StFeatures=MuFeatures.dropna()
ZtFeatures=MuFeatures.dropna()
```

```python
from scipy.stats import kendalltau,spearmanr,pearsonr

def check_direction2(r,y):
    pos = (r>0) & (y>0)
    neg = (r<=0) & (y<=0)
    mask = pos | neg
    return mask.sum()/mask.count()

Kendall = [kendalltau(MuFeatures['R'],MuFeatures[c])[0] for c in MuFeatures.columns[1:]]
Spearman = [spearmanr(MuFeatures['R'],MuFeatures[c])[0] for c in MuFeatures.columns[1:]]
Pearson = [pearsonr(MuFeatures['R'],MuFeatures[c])[0] for c in MuFeatures.columns[1:]]

relation = pd.DataFrame()
relation['Kendall'] = Kendall
relation['Spearman'] = Spearman
relation['Pearson'] = Pearson
relation.index = MuFeatures.columns[1:]
```
